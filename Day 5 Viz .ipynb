### Python

```python
df_fbi = pd.read_excel("FBI_stack.xls", sheet_name= "Property Rate")
```

```python
df_fbi = df_fbi.rename(columns={"Burglary \nrate ":"Burglary rate", 
                       "Larceny-\ntheft rate ": "Larceny-theft rate", 
                       "Motor \nvehicle \ntheft \nrate ": "Motor theft rate"})
```

```python
df_fbi
```

```python
colors = ["#011638", "#7e2987", "ef2026"]
labels = ["Motor theft rate", "Burglary rate", "Larceny-theft rate"]
sns.set_style("white") 
plt.figure(figsize = (12, 6))
plt.stackplot(df_fbi["Year"],
              df_fbi["Motor theft rate"],
              df_fbi["Burglary rate"],
              df_fbi["Larceny-theft rate"],
              colors = colors,
              edgecolor = 'none')
plt.xticks(df_fbi["Year"], rotation = 45)
plt.legend(labels = labels, loc = "upper right") 
plt.xlabel("Year", fontsize = 13)
plt.ylabel("Rate", fontsize = 13)
plt.title("Property Rate: Larceny-Theft has highest rate in US (1997-2016)", fontsize = 14, weight = "bold")
sns.despine() # Remove top and right border of the chart.
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize = (12, 6))
df_fbi.plot.area(x = 'Year', y = ['Motor theft rate', "Burglary rate", "Larceny-theft rate"])
plt.xlabel("Year", fontsize = 13)
plt.ylabel("Rate", fontsize = 13)
plt.title("Property Rate: Larceny-Theft has highest rate in US (1997-2016)", fontsize = 14, weight = "bold")
ax = plt.gca()
handles, labels = ax.get_legend_handles_labels()
labels, handles = zip(*sorted(zip(labels, handles), reverse=True))
ax.legend(handles, labels)
plt.show()
```

```python
df_stock = pd.read_excel("Stocks.xlsx", sheet_name= "Data", skiprows=1, skipfooter=5)
df_stock['S&P 500'] = df_stock['S&P 500']*100 
df_stock['FTSE 100'] = df_stock['FTSE 100']*100
```

```python
labels = ["S&P 500", "FTSE 100"] 
plt.figure(figsize = (20, 8))
plt.plot(df_stock["Year"], df_stock["S&P 500"], color = 'midnightblue')
plt.plot(df_stock["Year"], df_stock["FTSE 100"], color  = 'red')
plt.title("S&P vs FTSE Returns (1986 - 2010)", fontsize = 14, fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```
